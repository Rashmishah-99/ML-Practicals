## Find the difference between Python list and NumPy

| **Python List**                                    | **NumPy Array**                                             |
| -------------------------------------------------- | ----------------------------------------------------------- |
| Built-in data structure in Python                  | Provided by the **NumPy** library                           |
| Can store **different data types** (heterogeneous) | Stores **same data type** (homogeneous)                     |
| Slower for numerical operations                    | Much **faster** for numerical computations                  |
| Uses more memory                                   | **Memory efficient**                                        |
| Supports basic operations                          | Supports **advanced mathematical & statistical operations** |
| Element-wise operations need loops                 | Supports **vectorized operations** (no loops needed)        |
| Dynamic size (easy to append/remove)               | Fixed size (resizing is costly)                             |
| Suitable for general-purpose programming           | Ideal for **scientific computing & data analysis**          |
| No built-in support for multi-dimensional math     | Strong support for **multi-dimensional arrays**             |
| Example: `[1, 'a', 3.5]`                           | Example: `array([1, 2, 3])`                                 |

| ---------------------------------- | ------------------------------------------- |

| Uses more memory                   | **Memory efficient**                        |
| Element-wise operations need loops | Supports **direct element-wise operations** |
| Suitable for general programming   | Suitable for **scientific computing**       |


| **1D NumPy Array**                     | **Pandas Series**                     |
| -------------------------------------- | ------------------------------------- |
| Part of **NumPy** library              | Part of **Pandas** library            |
| Stores only values                     | Stores **values with labels (index)** |
| Index is always numeric (0,1,2â€¦)       | Custom index allowed                  |
| Faster for numerical computation       | Better for **data analysis**          |
| No handling of missing data by default | Supports **missing values (NaN)**     |
| Less descriptive                       | More **data-friendly and readable**   |

## Implement Boxplot, Scatter plot, line plot with suitable data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='price', data=df)
plt.title('Boxplot of Price by Class')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['price'], df['duration'], alpha=0.5)
plt.xlabel('Price')
plt.ylabel('Duration')
plt.title('Scatter Plot of Price vs Duration')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df.index, df['price'])
plt.xlabel('Index')
plt.ylabel('Price')
plt.title('Line Plot of Price over Index')
plt.show()
```
